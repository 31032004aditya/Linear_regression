```python
X_train = train_df.loc[:,'A':'U'].values
X_test = test_df.loc[:,'A':'U'].values
```

```python
Y_train = train_df.loc[:,'label'].values
Y_test = test_df.loc[:,'label'].values
```

```python
print('Shape of X_train:',X_train.shape)
print('Shape of Y_train:',Y_train.shape)
print('Shape of X_test:',X_test.shape)
print('Shape of Y_test:',Y_test.shape)
```

```python
Y_train = Y_train.reshape(Y_train.shape[0],1)
Y_test = Y_test.reshape(Y_test.shape[0],1)
print('Shape of Y_train:',Y_train.shape)
print('Shape of Y_test:',Y_test.shape)
```

```python
def linear_reg(X_train, Y_train, learning_rate, iteration):
    m = Y_train.size
    theta = np.zeros((X_train.shape[1],1))
    cost_list = []
    
    for i in range(iteration):
        
        Y_prediction = np.dot(X_train, theta)
        
        cost = (1/(2*m))*(np.sum(np.square(Y_prediction-Y_train)))
        
        d_theta = (1/m)*np.dot(X_train.T, Y_prediction - Y_train)
        
        theta = theta - learning_rate*d_theta
        
        cost_list.append(cost)
        
           # to print the cost for 10 times
#         if(i%(iteration/10) == 0):
#             print("Cost is :", cost)

        
    return theta, cost_list       
```

```python
iteration = 50000
learning_rate = 0.000001
theta, cost_list = linear_reg(X_train, Y_train, learning_rate = learning_rate, iteration = iteration)
```

```python
rng = np.arange(0, iteration)
plt.plot(rng, cost_list)
plt.show()
```

```python
Y_prediction = np.dot(X_test, theta)
print(Y_prediction - Y_test)
```
